```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Assuming df is your DataFrame
X = df[['year']]
y = df['per capita income (US$)']

# Change the order of the polynomial (degree)
degree =4 # For a quadratic regression
poly = PolynomialFeatures(degree)
poly_reg = make_pipeline(poly, LinearRegression())

# Fit the model
poly_reg.fit(X, y)

# Make predictions
predictions = poly_reg.predict(X)

# Plotting the results
plt.scatter(df['year'], df['per capita income (US$)'], color='blue')
plt.plot(df['year'], predictions, color='red')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title(f'Scatter Plot of Year vs Per Capita Income with Polynomial Degree {degree}')
plt.show()
```

```python
# Assuming reg is your regression model and df is your DataFrame
# You need to use the features and target variable that you used for fitting the model
X = df[['year']]
y = df['per capita income (US$)']

# Calculating the accuracy (R^2 score)
accuracy = poly_reg.score(X, y)
print(f'Accuracy (R^2 score): {accuracy}')
```
